```python
# Install NLTK (run this in terminal, not inside Python)
# pip install nltk

import nltk
from nltk.chat.util import Chat, reflections

# Download necessary NLTK data
nltk.download('punkt')
nltk.download('wordnet')

# Fixed patterns list
patterns = [
    (r'hi|hello|hey', ['Hi there!', 'Hello! How can I help you with books today?', 'Hey!']),
    (r'how are you\??', ["I'm good, thank you!", "I'm doing well, thanks for asking."]),
    (r'what is your name\??', ['You can call me Booksy.', 'I go by the name Booksy.']),
    (r'(.*) your name\??', ['You can call me Booksy.', 'I go by the name Booksy.']),
    (r'what can you do\??', ['I can answer questions about books and authors!', 'Ask me anything about books and I’ll do my best to help.']),
    (r'who is the author of "To Kill a Mockingbird"\??', ['The author of "To Kill a Mockingbird" is Harper Lee.']),
    (r'what is the book "1984" about\??', ['"1984" is about a dystopian future under a totalitarian regime led by Big Brother.']),
    (r'when was the book "The Great Gatsby" published\??', ['"The Great Gatsby" was published in 1925.']),
    (r'what genre is the book Moby-Dick\??', ['"Moby-Dick" falls under Adventure and Classic Literature.']),
    (r'what is the climax of Hamlet\??', ['The climax of "Hamlet" is the play-within-a-play where Hamlet confirms Claudius’s guilt.']),
    (r'how many pages are in the book Pride and Prejudice\??', ['"Pride and Prejudice" has approximately 432 pages.']),
    (r'what is the ISBN of the book "The Catcher in the Rye"\??', ['The ISBN is 978-0-316-76948-0.']),
    (r'can you recommend a book\??', ['I recommend "To Kill a Mockingbird" by Harper Lee.']),
    (r'who wrote The Lord of the Rings\??', ['J.R.R. Tolkien wrote The Lord of the Rings.']),
    (r'what are the main themes in Brave New World\??', ['Themes include dystopia, technology, and loss of individuality.']),
    (r'who is the protagonist of The Hobbit\??', ['Bilbo Baggins is the protagonist of The Hobbit.']),
    (r'who are the characters in The Chronicles of Narnia\??', ['Lucy, Edmund, Susan, and Peter Pevensie.']),
    (r'where does "Jane Eyre" take place\??', ['"Jane Eyre" is set primarily in northern England.']),
    (r'what is the setting of Wuthering Heights\??', ['The Yorkshire moors in England.']),
    (r'what book won the Pulitzer Prize in 2020\??', ['"The Nickel Boys" won the 2020 Pulitzer Prize for Fiction.']),
    (r'what is the plot of "The Da Vinci Code"\??', ['It follows a symbologist uncovering secrets about the Holy Grail.']),
    (r'how does Gone with the Wind end\??', ['Scarlett vows to rebuild her life even after losing Rhett Butler.']),
    (r'who is the antagonist in "Dracula"\??', ['Count Dracula is the antagonist.']),
    (r'what book did J\.K\. Rowling write\??', ['She wrote "Harry Potter and the Sorcerer’s Stone".']),
    (r'how many chapters are in The Hunger Games\??', ['"The Hunger Games" has 27 chapters.']),
    (r'what is the review of "The Road"\??', ['Reviews praise its emotional post-apocalyptic storytelling.']),
    (r'who is the illustrator of "Where the Wild Things Are"\??', ['Maurice Sendak illustrated it.']),
    (r'what is George Orwell\'s most popular work\??', ['His most popular book is "1984".']),
    (r'what is "The Girl with the Dragon Tattoo" famous for\??', ['For its gripping mystery and complex characters.']),
    (r'what is "Little Women" acclaimed for\??', ['Its depiction of family life and strong female characters.']),
    (r'who is the publisher of "Murder on the Orient Express"\??', ['Collins Crime Club published it.']),
    (r'what is the synopsis of "Life of Pi"\??', ['A boy survives at sea on a lifeboat with a Bengal tiger.']),
    (r'where can I buy "The Fault in Our Stars"\??', ['Amazon or Barnes & Noble.']),
    (r'what are reviews of "Sapiens" like\??', ['They praise its engaging look at human history.']),
    (r'who reviewed "The Alchemist"\??', ['Reviewers include critics from The New York Times and The Guardian.']),
    (r'what are the sales figures for The Shining\??', ['Over 2 million copies sold worldwide.']),
    (r'what book should I read next\??', ['Try "The Goldfinch" by Donna Tartt.']),
    (r'what is "The Bell Jar" about\??', ['It’s about a young woman’s struggle with mental illness.']),
    (r'who is the author of Slaughterhouse-Five\??', ['Kurt Vonnegut wrote it.']),
    (r'who are the main characters in The Outsiders\??', ['Ponyboy Curtis, Johnny Cade, and Dallas Winston.']),
]

# Create a chatbot
chatbot = Chat(patterns, reflections)

print("Hello! I'm Booksy. How can I help you today?")

while True:
    user_input = input("You: ").lower()

    if user_input in ['bye', 'later', 'see you']:
        if user_input == 'bye':
            print("Booksy: Bye! Take care!")
        elif user_input == 'later':
            print("Booksy: See you later! Stay awesome!")
        else:
            print("Booksy: See you soon! Take care!")
        break

    response = chatbot.respond(user_input)

    # Fallback if no pattern matched
    if response is None:
        response = "Sorry, I’m not sure about that. Try asking something about books or authors!"

    print("Booksy:", response)
```
